```python
def classify(time):
    if time<=1950:
        return 'Vintage house'
    elif time<=1980:
        return 'old house'
    elif time<=2000:
        return 'new house'
    else:
        return 'no house'
```

```python
data['House Type']=data['YearBuilt'].apply(classify)
```

```python
#classify a house in very big,big and small according to the squarefoot
```

```python
def tell(Feet):
    if Feet<=5000:
        return 'Very big'
    elif Feet<=3000:
        return 'big'
    else:
        return 'small'
```

```python
data['House size']=data['SquareFeet'].apply(tell)
```

```python
data['new price']=data['Price']*1.1
```
