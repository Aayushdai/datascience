```python
# IQR method
q1=data['Mileage'].quantile(0.25)
q3=data['Mileage'].quantile(0.75)
```

```python
iqr=q3-q1
```

```python
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
```

```python
sns.kdeplot(x='Mileage_clip',data=data)
plt.axvline(lower_bound,color='green',linestyle='--')
plt.axvline(upper_bound,color='red',linestyle='--')
```

```python
data['Mileage_iqr']=np.clip(data['Mileage'],lower_bound,upper_bound)
```

```python
sns.kdeplot(x='Mileage_iqr',data=data)
```
