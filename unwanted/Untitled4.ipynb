```python
import pandas as pd
data=pd.read_csv(r'cars_ready_to_eda.csv')
```

```python
# count the cars running in different fuel types
```

```python
data['Fuel_Type'].nunique()
```

```python
data['Fuel_Type'].unique()
```

```python
data.shape
```

```python
sns.countplot(x='Fuel_Type',data=data,palette='cool')
```

# countplot,barplot,lineplot,scatterplot,boxplot,vailin plot,ideplot,histgram,hratmap,pair plot,pie chart

```python
sns.barplot(x='Fuel_Type',y='Mileage',data=data,palette='colorblind')
```

# calculate the avg of power wrt location

```python
plt.title("Location and Power Distribution")
sns.barplot(x='Location',y='Power',data=data,palette='colorblind')
plt.grid()
plt.xticks(rotation=90)
```

```python
count the no of cars according to brand
```

```python
plt.title("No of cars and it's brand")
sns.barplot(x='Location',y='brand',data=data,palette='colorblind')
plt.grid()
plt.xticks(rotation=90)
```

```python
sns.lineplot(x='Location',y='Mileage', marker='D',ci=None,data=data,label='Avg Mileage')
sns.lineplot(x='Location',y='Power',data=data,label='avg power')
plt.legend()
plt.xticks(rotation=90)
```

```python
sns.scatterplot(y='Power',x='Mileage',hue='Fuel_Type',data=data)
```

# Strip plot

```python
sns.stripplot(x='Fuel_Type',y='Mileage',hue='Owner_Type',data=data,)
```

# BOX PLOT

```python
sns.boxplot(x='Mileage',data=data)
```

```python
sns.boxplot(y='Mileage',data=data)
```

```python
sns.boxplot(y='Power',data=data,hue='Fuel_Type')
```

```python
# Kde Plot
sns.kdeplot(x='Power',hue='Fuel_Type',data=data)
```

```python
sns.histplot(x='brand',kde='True',hue='Fuel_Type',data=data)
plt.xticks(rotation=90)
```
